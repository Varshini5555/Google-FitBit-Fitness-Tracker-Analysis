```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/dailyActivity_merged.csv")
data.head()
```

Checking for null values in the dataset

```python
data.isnull().sum()
```

```python
data.info()
```

Converting the datatype of "ActivitiyDate" from object to datetime

```python
data['ActivityDate']=pd.to_datetime(data['ActivityDate'])
```

```python
data.info()
```

Generating variable for unique ids of users

```python
unique_id=pd.unique(data["Id"])
len(unique_id)
```

```python
new_cols = ['Id', 'ActivityDate', 'DayOfTheWeek', 'TotalSteps', 'TotalDistance', 'TrackerDistance', 'LoggedActivitiesDistance', 'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes', 'TotalExerciseMinutes', 'TotalExerciseHours', 'Calories']
data1 = data.reindex(columns=new_cols)
data1.head(5)
```

```python
data1['DayOfTheWeek']=data1['ActivityDate'].dt.day_name()
```

```python
data1.head()
```

```python
data1['TotalExerciseMinutes']=data1['VeryActiveMinutes']+data1['FairlyActiveMinutes']+data1['LightlyActiveMinutes']+data1['SedentaryMinutes']
data1['TotalExerciseMinutes'].head()
```

```python
data1['TotalExerciseHours']=data1['TotalExerciseMinutes']/60
data1['TotalExerciseHours'].head()
```
