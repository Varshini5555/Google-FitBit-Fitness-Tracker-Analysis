```python
very_active_mins =data["VeryActiveMinutes"].sum()
fairly_active_mins =data["FairlyActiveMinutes"].sum()
lightly_active_mins = data["LightlyActiveMinutes"].sum()
sedentary_mins = data["SedentaryMinutes"].sum()

slices = [very_active_mins, fairly_active_mins, lightly_active_mins, sedentary_mins]
labels = ["Very active minutes", "Fairly active minutes", "Lightly active minutes", "Sedentary minutes"]
colours = ["lightcoral", "green", "skyblue", "greenyellow"]
plt.style.use("default")
plt.pie(slices, labels = labels,
        colors = colours, wedgeprops = {"edgecolor": "black"}, autopct = "%1.1f%%")
plt.title("Percentage of Activity in Minutes")
plt.tight_layout()
plt.show()
```
